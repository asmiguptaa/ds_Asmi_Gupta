```python
trades = pd.read_csv('historical_data.csv')
sentiment = pd.read_csv('fear_greed_index.csv')
```

```python
trades.head()
sentiment.head()
```

```python
trades.columns
```

```python
trades.columns = (
    trades.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)

```

```python
trades.columns
```

```python
trades['timestamp'] = pd.to_datetime(trades['timestamp'])
trades['date'] = trades['timestamp'].dt.date
```

```python
trades[['timestamp', 'date']].head()
```

```python
sentiment.columns = sentiment.columns.str.strip().str.lower()
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date
```

```python
merged_df = pd.merge(trades, sentiment, on='date', how='inner')
```

```python
trades = trades.dropna(subset=[
    'execution_price',
    'size_usd',
    'closed_pnl',
    'leverage' if 'leverage' in trades.columns else 'size_usd'
])
```

```python
trades.columns
```

```python
trades['is_profitable'] = trades['closed_pnl'] > 0
```

```python
trades['trade_volume'] = trades['execution_price'] * trades['size_tokens']
```

```python
daily_metrics = trades.groupby('date').agg(
    total_volume_usd = ('trade_volume', 'sum'),
    total_pnl = ('closed_pnl', 'sum'),
    avg_trade_size = ('size_usd', 'mean'),
    profitable_trade_ratio = ('is_profitable', 'mean'),
    active_traders = ('account', 'nunique')
).reset_index()
```

```python
merged_df = pd.merge(
    daily_metrics,
    sentiment[['date', 'classification']],
    on='date',
    how='inner'
)
```

```python
import os

os.makedirs('csv_files', exist_ok=True)
os.makedirs('outputs', exist_ok=True)
```

```python
merged_df.to_csv('csv_files/merged_daily_metrics.csv', index=False)
daily_metrics.to_csv('csv_files/daily_metrics.csv', index=False)
```

```python
os.listdir('csv_files')
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='classification', y='total_pnl', data=merged_df)
plt.title('Daily PnL vs Market Sentiment')
plt.tight_layout()
plt.savefig('outputs/pnl_vs_sentiment.png')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='classification', y='profitable_trade_ratio', data=merged_df)
plt.title('Profitable Trade Ratio by Sentiment')
plt.tight_layout()
plt.savefig('outputs/profit_ratio.png')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='classification', y='total_volume_usd', data=merged_df)
plt.title('Trading Volume by Market Sentiment')
plt.tight_layout()
plt.savefig('outputs/volume_by_sentiment.png')
plt.show()
```
